```python
# =========================
# Cell 1.1: Imports & Logging Configuration
# =========================

import numpy as np
import matplotlib.pyplot as plt
import logging
import sys
import os
import re  # used for extracting snapshot index from filenames

# Configure logging
logging.basicConfig(
    level=logging.INFO,  # adjust as needed
    format='%(asctime)s [%(levelname)s] %(message)s',
    handlers=[logging.StreamHandler(sys.stdout)]
)

logging.info("Notebook started. Libraries imported successfully.")

# Example constants (keep if needed)
G_kpcGyrMsun      = 4.499e-6  # kpc^3 / Gyr^2 / Msun
M31_halo_mass     = 1.5e12    # Msun
M31_scale_radius  = 60.0      # kpc

print("Cell 1.1 completed: Libraries and logging configured.")
```

```python
# =========================
# Cell 1.2: User-Configurable Parameters & Filenames
# =========================

# Directories for snapshot files
M31_DIRECTORY = "M31/"  
M33_DIRECTORY = "M33/"  

# Tidal / mass-loss config
ROUTER_LIMIT   = 30.0   # maximum radius for bound mass check (kpc)

# Surface density profile config
NBINS_PROFILES = 30     
RMAX_PROFILES  = 30.0   

# M31 halo params (if using Hernquist approximation)
M31_HALO_MASS    = 1.5e12
M31_SCALE_RADIUS = 60.0

# Data logging or output options
SAVE_INTERMEDIATE_RESULTS = True  
PLOTS_DIRECTORY           = "plots/"
RESULTS_DIRECTORY         = "results/"

print("Cell 1.2 completed: User-configurable parameters and filenames set.")
print(f"M31 files assumed to be in: {M31_DIRECTORY}")
print(f"M33 files assumed to be in: {M33_DIRECTORY}")
print(f"Router limit for bound mass: {ROUTER_LIMIT} kpc")
print(f"Surface density profile: {NBINS_PROFILES} radial bins, rmax={RMAX_PROFILES} kpc")
```

```python
# =========================
# Cell 2.1 — File Parsing Functions
# =========================

def parse_galaxy_file(filename):
    """
    Reads a snapshot file (e.g. M31_000.txt, M33_010.txt) and returns:
      - time_val (float): Time in Gyr from file header
      - total_particles (int)
      - data_tuple: (ptype, pmass, x, y, z, vx, vy, vz)
    """
    time_val = None
    total_particles = None

    ptype_list = []
    mass_list  = []
    x_list     = []
    y_list     = []
    z_list     = []
    vx_list    = []
    vy_list    = []
    vz_list    = []

    with open(filename, 'r') as f:
        lines = f.readlines()

    data_start_idx = None

    # Parse header lines
    for i, line in enumerate(lines):
        line_stripped = line.strip()
        if line_stripped.startswith("Time"):
            parts = line_stripped.split()
            time_val = float(parts[1])
        elif line_stripped.startswith("Total"):
            parts = line_stripped.split()
            total_particles = int(parts[1])
        elif line_stripped.startswith("#type"):
            data_start_idx = i + 1
            break

    if data_start_idx is None:
        logging.warning(f"No data header ('#type') found in {filename}")
        return None, None, (None,)*8

    # Parse data rows
    for line in lines[data_start_idx:]:
        line_stripped = line.strip()
        if not line_stripped:
            continue
        parts = line_stripped.split()
        if len(parts) < 8:
            continue

        ptype = float(parts[0])
        pmass = float(parts[1])
        px    = float(parts[2])
        py    = float(parts[3])
        pz    = float(parts[4])
        pvx   = float(parts[5])
        pvy   = float(parts[6])
        pvz   = float(parts[7])

        ptype_list.append(ptype)
        mass_list.append(pmass)
        x_list.append(px)
        y_list.append(py)
        z_list.append(pz)
        vx_list.append(pvx)
        vy_list.append(pvy)
        vz_list.append(pvz)

    ptype_array = np.array(ptype_list)
    mass_array  = np.array(mass_list)
    x_array     = np.array(x_list)
    y_array     = np.array(y_list)
    z_array     = np.array(z_list)
    vx_array    = np.array(vx_list)
    vy_array    = np.array(vy_list)
    vz_array    = np.array(vz_list)

    if total_particles and (len(ptype_array) != total_particles):
        logging.warning(f"Mismatch in particle count for {filename}: "
                        f"header={total_particles}, actual={len(ptype_array)}")

    logging.info(f"Parsed file: {filename} | time={time_val} Gyr | total={len(ptype_array)}")
    return time_val, len(ptype_array), (ptype_array, mass_array, 
                                        x_array, y_array, z_array, 
                                        vx_array, vy_array, vz_array)


def get_snapshot_index(filename):
    """
    Extracts snapshot index if filename has pattern like 'M31_003.txt'.
    Returns int or None if not found.
    """
    match = re.search(r'_([0-9]+)\.txt', filename)
    if match:
        return int(match.group(1))
    return None

print("Cell 2.1 completed: parse_galaxy_file() and get_snapshot_index() defined.")
```

```python
# =========================
# Cell 2.2 — Center of Mass & Basic Math
# =========================

def center_of_mass(x, y, z, mass):
    """
    Returns the (xcom, ycom, zcom) for the given arrays and mass.
    """
    M_total = np.sum(mass)
    if M_total == 0:
        logging.warning("center_of_mass: total mass is zero, returning (0,0,0).")
        return 0.0, 0.0, 0.0

    xcom = np.sum(x * mass) / M_total
    ycom = np.sum(y * mass) / M_total
    zcom = np.sum(z * mass) / M_total
    return xcom, ycom, zcom


def hernquist_enclosed_mass(r, M_halo=1.5e12, a=60.0):
    """
    Hernquist profile: M(r) = M_halo * [ r^2 / (r + a)^2 ].
    """
    r = np.atleast_1d(r)
    M_enc = M_halo * (r**2 / (r + a)**2)
    return M_enc[0] if len(M_enc) == 1 else M_enc


def distance_3d(x1, y1, z1, x2, y2, z2):
    """3D distance between two points."""
    dx, dy, dz = (x2 - x1), (y2 - y1), (z2 - z1)
    return np.sqrt(dx*dx + dy*dy + dz*dz)


def vector_magnitude(x, y, z):
    """
    Returns sqrt(x^2 + y^2 + z^2).
    Works element-wise if x,y,z are arrays.
    """
    return np.sqrt(x*x + y*y + z*z)

print("Cell 2.2 completed: center_of_mass(), hernquist_enclosed_mass(), etc. defined.")
```

```python
# =========================
# Cell 2.3 — Diagnostic Test: Single Snapshot
# =========================

def diagnostic_test_single_snapshot(snapshot_index=0):
    """
    Quick parse for M31 and M33 at the same snapshot index, then prints
    time, particle counts, COM, and total mass.
    """
    file_m31 = f"{M31_DIRECTORY}M31_{snapshot_index:03d}.txt"
    file_m33 = f"{M33_DIRECTORY}M33_{snapshot_index:03d}.txt"

    if not (os.path.exists(file_m31) and os.path.exists(file_m33)):
        logging.warning(f"Missing M31/M33 snapshot {snapshot_index:03d}")
        return

    # M31
    t_m31, ct_m31, (ptype_m31, mass_m31, x_m31, y_m31, z_m31,
                    vx_m31, vy_m31, vz_m31) = parse_galaxy_file(file_m31)
    # M33
    t_m33, ct_m33, (ptype_m33, mass_m33, x_m33, y_m33, z_m33,
                    vx_m33, vy_m33, vz_m33) = parse_galaxy_file(file_m33)

    # Basic info
    print(f"\n=== Diagnostic Test for Snap Index {snapshot_index:03d} ===")
    print(f"M31: time={t_m31} Gyr, total_particles={ct_m31}")
    print(f"M33: time={t_m33} Gyr, total_particles={ct_m33}")

    # COM
    xcom_m31, ycom_m31, zcom_m31 = center_of_mass(x_m31, y_m31, z_m31, mass_m31)
    xcom_m33, ycom_m33, zcom_m33 = center_of_mass(x_m33, y_m33, z_m33, mass_m33)

    print(f"M31 COM: ({xcom_m31:.3f}, {ycom_m31:.3f}, {zcom_m31:.3f}) kpc")
    print(f"M33 COM: ({xcom_m33:.3f}, {ycom_m33:.3f}, {zcom_m33:.3f}) kpc")

    # Total mass (in Msun), since snapshot mass is in 1e10 Msun
    m31_total = mass_m31.sum() * 1e10
    m33_total = mass_m33.sum() * 1e10
    print(f"M31 total mass ~ {m31_total:.2e} Msun")
    print(f"M33 total mass ~ {m33_total:.2e} Msun")

    # Check if times differ significantly
    if abs(t_m31 - t_m33) > 1e-3:
        logging.warning("Times for M31 vs. M33 differ > 1e-3 Gyr in the same snapshot index!")

# Test it by default for snapshot 0
diagnostic_test_single_snapshot(0)

print("Cell 2.3 completed: single snapshot diagnostic ran.")
```

```python
import glob as glob

# ==============================
# Cell 3.1.1 — Compute Jacobi Radius for All Snapshots
# ==============================
def compute_jacobi_all_snaps(save_results=SAVE_INTERMEDIATE_RESULTS):
    """
    Computes M33's Jacobi radius at each snapshot by:
      1) Matching M31 & M33 snapshots
      2) Parsing & finding COM
      3) R = distance between centers
      4) M_enclosed from M31's Hernquist
      5) R_j = R * (M33 / (2*Mhost_enc))^(1/3)

    Returns times_jacobi, rjacobi_arr.
    Saves to .npy if save_results=True.
    """
    m31_pattern = os.path.join(M31_DIRECTORY, "M31_*.txt")
    m33_pattern = os.path.join(M33_DIRECTORY, "M33_*.txt")

    m31_files = sorted(glob.glob(m31_pattern))
    m33_files = sorted(glob.glob(m33_pattern))

    def get_snapidx(fname):
        return get_snapshot_index(fname)

    # Build dict for M31, M33 => {snapshot_index: filename}
    m31_dict = {}
    for f in m31_files:
        idx = get_snapidx(f)
        if idx is not None:
            m31_dict[idx] = f

    m33_dict = {}
    for f in m33_files:
        idx = get_snapidx(f)
        if idx is not None:
            m33_dict[idx] = f

    # Find shared snapshots
    common_indices = sorted(set(m31_dict.keys()) & set(m33_dict.keys()))
    if not common_indices:
        logging.warning("No common M31/M33 snapshots found!")
        return np.array([]), np.array([])

    times_list = []
    rj_list = []

    # Main loop
    for snap_idx in common_indices:
        # Parse M31
        file_m31 = m31_dict[snap_idx]
        t_m31, _, data_m31 = parse_galaxy_file(file_m31)
        (ptype_m31, mass_m31, x_m31, y_m31, z_m31,
         vx_m31, vy_m31, vz_m31) = data_m31

        # Parse M33
        file_m33 = m33_dict[snap_idx]
        t_m33, _, data_m33 = parse_galaxy_file(file_m33)
        (ptype_m33, mass_m33, x_m33, y_m33, z_m33,
         vx_m33, vy_m33, vz_m33) = data_m33

        # Time for that snapshot
        t_val = 0.5*(t_m31 + t_m33)

        # COM for M31 & M33
        xcom_m31, ycom_m31, zcom_m31 = center_of_mass(x_m31, y_m31, z_m31, mass_m31)
        xcom_m33, ycom_m33, zcom_m33 = center_of_mass(x_m33, y_m33, z_m33, mass_m33)

        # Distance
        dx = xcom_m33 - xcom_m31
        dy = ycom_m33 - ycom_m31
        dz = zcom_m33 - zcom_m31
        R  = np.sqrt(dx*dx + dy*dy + dz*dz)

        # M31 enclosed mass
        Mhost_enc = hernquist_enclosed_mass(R, M_halo=M31_HALO_MASS, a=M31_SCALE_RADIUS)

        # M33 total mass (Msun, file mass is in 1e10)
        M33_sat_mass = np.sum(mass_m33) * 1e10
        if Mhost_enc <= 0:
            logging.warning(f"Bad Mhost_enc for snap {snap_idx}, skipping.")
            continue

        # Jacobi radius
        R_j = R * ((M33_sat_mass / (2.0 * Mhost_enc)) ** (1.0/3.0))
        logging.info(f"Snap {snap_idx:03d} time={t_val:.2f} Gyr | R_j={R_j:.2f} kpc")

        times_list.append(t_val)
        rj_list.append(R_j)

    # Convert to arrays & sort by time
    times_jacobi = np.array(times_list)
    rjacobi_arr  = np.array(rj_list)
    idx_sort = np.argsort(times_jacobi)
    times_jacobi = times_jacobi[idx_sort]
    rjacobi_arr  = rjacobi_arr[idx_sort]

    # Optionally save
    if save_results and len(times_jacobi) > 0:
        os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(RESULTS_DIRECTORY, "jacobi_radius_vs_time.npy")
        np.save(outfile, np.column_stack((times_jacobi, rjacobi_arr)))
        logging.info(f"Jacobi radius results saved to {outfile}")

    return times_jacobi, rjacobi_arr


# Run it and store results
times_jacobi, rjacobi_arr = compute_jacobi_all_snaps()
print("Jacobi radius computed for all snapshots.")
```

```python
# ==============================
# Cell 3.1.2 — Plot Jacobi Radius vs. Time
# ==============================
def plot_jacobi_radius(times_jacobi, rjacobi_arr, outfig="jacobi_radius_over_time.png"):
    """
    Plots R_jacobi vs. time.
    """
    if len(times_jacobi) == 0:
        print("No Jacobi radius data to plot.")
        return

    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(times_jacobi, rjacobi_arr, '-', label='Jacobi Radius')

    ax.set_xlabel("Time (Gyr)")
    ax.set_ylabel("R_j (kpc)")
    ax.set_title("Jacobi Radius vs. Time")
    ax.grid(True)
    ax.legend()

    if outfig:
        os.makedirs(PLOTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(PLOTS_DIRECTORY, outfig)
        plt.tight_layout()
        plt.savefig(outfile, dpi=150)
        print(f"Jacobi radius plot saved to {outfile}")

    plt.show()


# Create the plot
plot_jacobi_radius(times_jacobi, rjacobi_arr)
print("Jacobi radius vs. time plot generated.")
```

```python
# ==============================
# Cell 3.2.1 — Compute M33 Bound Mass Over Time
# ==============================
def compute_bound_mass_all_snaps(times_jacobi, rjacobi_arr, 
                                 router_limit=ROUTER_LIMIT,
                                 save_results=SAVE_INTERMEDIATE_RESULTS):
    """
    For each snapshot, sets R_cut = min(R_j, router_limit),
    sums M33 mass inside R_cut, returns time, Mbound arrays.
    """
    m33_pattern = os.path.join(M33_DIRECTORY, "M33_*.txt")
    m33_files   = sorted(glob.glob(m33_pattern))

    # Build a dict for M33: {snapshot_idx: filename}
    m33_dict = {}
    for f in m33_files:
        idx = get_snapshot_index(f)
        if idx is not None:
            m33_dict[idx] = f

    if len(times_jacobi) == 0 or len(rjacobi_arr) == 0:
        logging.warning("compute_bound_mass_all_snaps: No Jacobi radius data provided.")
        return np.array([]), np.array([])

    times_bound = []
    Mbound_arr  = []

    m33_snap_indices = sorted(m33_dict.keys())
    if len(m33_snap_indices) != len(rjacobi_arr):
        logging.warning("Mismatch between M33 snapshot count and rjacobi_arr length.")

    n_common = min(len(m33_snap_indices), len(rjacobi_arr))
    for i in range(n_common):
        snap_idx = m33_snap_indices[i]
        file_m33 = m33_dict[snap_idx]

        t_val = times_jacobi[i]
        R_j   = rjacobi_arr[i]

        # Parse M33
        t_m33, _, data_m33 = parse_galaxy_file(file_m33)
        (ptype_m33, mass_m33, x_m33, y_m33, z_m33,
         vx_m33, vy_m33, vz_m33) = data_m33

        # COM shift
        xcom_m33, ycom_m33, zcom_m33 = center_of_mass(x_m33, y_m33, z_m33, mass_m33)
        x_shift = x_m33 - xcom_m33
        y_shift = y_m33 - ycom_m33
        z_shift = z_m33 - zcom_m33
        r_m33   = np.sqrt(x_shift**2 + y_shift**2 + z_shift**2)

        R_cut = min(R_j, router_limit)
        inside = (r_m33 <= R_cut)
        M_bound = np.sum(mass_m33[inside]) * 1e10  # Msun

        times_bound.append(t_val)
        Mbound_arr.append(M_bound)
        logging.info(f"Snap {snap_idx:03d}, time={t_val:.2f} Gyr, R_cut={R_cut:.2f} kpc, Bound={M_bound:.2e} Msun")

    # Convert & sort by time
    times_bound = np.array(times_bound)
    Mbound_arr  = np.array(Mbound_arr)
    idx_sort    = np.argsort(times_bound)
    times_bound = times_bound[idx_sort]
    Mbound_arr  = Mbound_arr[idx_sort]

    if save_results and len(times_bound) > 0:
        os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(RESULTS_DIRECTORY, "m33_bound_mass_vs_time.npy")
        np.save(outfile, np.column_stack((times_bound, Mbound_arr)))
        logging.info(f"Bound mass results saved to {outfile}")

    return times_bound, Mbound_arr

# Run it
times_bound, Mbound_arr = compute_bound_mass_all_snaps(times_jacobi, rjacobi_arr)
print("M33 bound mass computed & stored for all snapshots.")
```

```python
# ==============================
# Cell 3.2.2 — Plot Bound Mass Timeseries
# ==============================
def plot_m33_bound_mass(times_bound, Mbound_arr, outfig="M33_bound_mass_vs_time.png"):
    """
    Plots M33 bound mass vs. time, plus a quick annotation for total % mass lost.
    """
    if len(times_bound) == 0:
        print("No bound mass data to plot.")
        return

    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(times_bound, Mbound_arr, '-', color='purple', label='M33 Bound Mass')
    ax.set_xlabel("Time (Gyr)")
    ax.set_ylabel("Mass (Msun)")
    ax.set_title("M33 Bound Mass vs. Time")
    ax.grid(True)
    ax.legend()

    initial_mass = Mbound_arr[0]
    final_mass   = Mbound_arr[-1]
    if initial_mass > 0:
        loss_pct = 100 * (initial_mass - final_mass) / initial_mass
    else:
        loss_pct = 0

    text_str = (f"Initial: {initial_mass:.2e}\n"
                f"Final:   {final_mass:.2e}\n"
                f"Lost ~ {loss_pct:.1f}%")
    ax.text(0.05, 0.8, text_str, transform=ax.transAxes,
            fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

    if outfig:
        os.makedirs(PLOTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(PLOTS_DIRECTORY, outfig)
        plt.tight_layout()
        plt.savefig(outfile, dpi=150)
        print(f"Bound mass plot saved to {outfile}")

    plt.show()

    print("Bound Mass Discussion:")
    print(f" - Initial: {initial_mass:.2e} Msun")
    print(f" - Final:   {final_mass:.2e} Msun")
    print(f" - Lost ~ {loss_pct:.1f}% of initial.\n")

# Plot results
plot_m33_bound_mass(times_bound, Mbound_arr)
print("M33 bound mass timeseries plot generated.")
```

```python
# =========================================
# Cell 4.1.1 — Surface Density Profile + Single Snapshot Demo
# =========================================

def surface_density_profile(x, y, mass, nbins=NBINS_PROFILES, rmax=RMAX_PROFILES):
    """
    Compute a radial surface density profile: 
      - Bins from 0..rmax in nbins annuli.
      - Returns r_mid, Sigma in Msun/kpc^2.
    """
    mass_in_Msun = mass * 1e10  # if mass is in 1e10 Msun
    R = np.sqrt(x*x + y*y)

    edges = np.linspace(0, rmax, nbins+1)
    r_mid = 0.5*(edges[:-1] + edges[1:])
    mass_in_bin = np.zeros(nbins)

    for i in range(nbins):
        mask = (R >= edges[i]) & (R < edges[i+1])
        mass_in_bin[i] = mass_in_Msun[mask].sum()

    area_bins = np.pi*(edges[1:]**2 - edges[:-1]**2)
    Sigma = mass_in_bin / area_bins
    return r_mid, Sigma


def demo_surface_density_single_snapshot(snap_index=0):
    """
    Parse M33_<snap_index>.txt, center on M33, compute (r_mid, Sigma),
    log-log plot for quick inspection.
    """
    file_m33 = f"{M33_DIRECTORY}M33_{snap_index:03d}.txt"
    t_m33, _, data_m33 = parse_galaxy_file(file_m33)
    (ptype_m33, mass_m33, x_m33, y_m33, z_m33,
     vx_m33, vy_m33, vz_m33) = data_m33

    # Center on M33
    xcom_m33, ycom_m33, zcom_m33 = center_of_mass(x_m33, y_m33, z_m33, mass_m33)
    x_shift = x_m33 - xcom_m33
    y_shift = y_m33 - ycom_m33

    # Surface density
    r_mid, Sigma = surface_density_profile(x_shift, y_shift, mass_m33)

    # Plot
    fig, ax = plt.subplots(figsize=(7,5))
    ax.loglog(r_mid, Sigma, '-', label='Surface Density')
    ax.set_xlabel("R (kpc)")
    ax.set_ylabel(r"$\Sigma(R)$ [Msun/kpc$^2$]")
    ax.set_title(f"M33_{snap_index:03d} at t={t_m33:.1f} Gyr")
    ax.legend()
    plt.tight_layout()

    # Save
    os.makedirs(PLOTS_DIRECTORY, exist_ok=True)
    outfig = os.path.join(PLOTS_DIRECTORY, f"M33_{snap_index:03d}_SurfaceDensityDemo.png")
    plt.savefig(outfig, dpi=150)
    print(f"Surface density demo plot saved -> {outfig}")
    plt.show()

    logging.info(f"demo_surface_density_single_snapshot snap={snap_index} => maxSigma={Sigma.max():.2e}")


# Example usage (can comment out if you don’t want to run automatically):
demo_surface_density_single_snapshot(0)
print("Cell 4.1.1 done: surface_density_profile() & single-snapshot demo.")
```

```python
# =========================================
# Cell 4.1.2 — Generate Surface Density Profiles for All Snapshots
# =========================================

def generate_all_surface_density_profiles(nbins=NBINS_PROFILES, rmax=RMAX_PROFILES,
                                          save_results=SAVE_INTERMEDIATE_RESULTS):
    """
    For each M33 snapshot:
      1) parse & center
      2) compute (r_mid, Sigma) with surface_density_profile
      3) store in a dict: profiles[snap_idx] = (time, r_mid, Sigma)

    Returns dict of snapshot_index -> (time[Gyr], r_mid, Sigma).
    """
    m33_files = sorted(glob.glob(f"{M33_DIRECTORY}M33_*.txt"))
    profiles = {}

    for f in m33_files:
        snap_idx = get_snapshot_index(f)
        if snap_idx is None:
            continue

        t_m33, _, data_m33 = parse_galaxy_file(f)
        (ptype, mass, x, y, z, vx, vy, vz) = data_m33

        xcom, ycom, zcom = center_of_mass(x, y, z, mass)
        x_shift = x - xcom
        y_shift = y - ycom

        r_mid, Sigma = surface_density_profile(x_shift, y_shift, mass,
                                               nbins=nbins, rmax=rmax)
        profiles[snap_idx] = (t_m33, r_mid, Sigma)

        if np.isnan(Sigma).any() or (Sigma.size > 0 and Sigma.max() > 1e12):
            logging.warning(f"Snapshot {snap_idx}: suspicious Sigma => max={Sigma.max():.2e}")

    if save_results and len(profiles) > 0:
        os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
        outfn = os.path.join(RESULTS_DIRECTORY, "all_surface_density_profiles.npz")

        savez_dict = {}
        for snap_idx, (t_val, r_arr, s_arr) in profiles.items():
            savez_dict[f"Snap{snap_idx}_time"]   = np.array([t_val])
            savez_dict[f"Snap{snap_idx}_rmid"]   = r_arr
            savez_dict[f"Snap{snap_idx}_Sigma"]  = s_arr
        np.savez(outfn, **savez_dict)
        logging.info(f"All surface density profiles saved -> {outfn}")

    return profiles

# Generate and store
all_profiles = generate_all_surface_density_profiles()
print(f"Cell 4.1.2 done: generated surface density profiles for {len(all_profiles)} snapshots.")
```

```python
import numpy as np
import logging
from scipy.optimize import curve_fit
# =========================================
# Cell 4.2.1 — Define Sersic/Exponential Fitting Routines
# =========================================

def sersic_profile(R, Sigma_e, Re, n):
    """
    Sigma(R) = Sigma_e * exp( -b_n [ (R/Re)^(1/n) - 1 ] ), 
    with b_n ~ 1.9992n - 0.3271.
    """
    # ensure array
    R = np.atleast_1d(R)
    b_n = max(1.9992*n - 0.3271, 1e-3)
    term = (R/Re)**(1./n) - 1
    return Sigma_e * np.exp(-b_n * term)


def exponential_profile(R, Sigma0, h):
    """Sigma(R) = Sigma0 * exp(-R/h)."""
    R = np.atleast_1d(R)
    return Sigma0 * np.exp(-R/h)


def fit_sersic(rvals, sigma_vals):
    """
    Fit Sersic to (rvals, sigma_vals). Return (Sigma_e, Re, n).
    If failing, returns (nan, nan, nan).
    """
    mask = (rvals>0) & (sigma_vals>0)
    r_fit = rvals[mask]
    s_fit = sigma_vals[mask]
    if len(r_fit) < 5:
        logging.warning("fit_sersic: insufficient data points.")
        return (np.nan, np.nan, np.nan)

    # guesses
    guess_Sigma_e = np.median(s_fit)
    guess_Re      = r_fit[len(r_fit)//2] if len(r_fit)>1 else 5.0
    guess_n       = 1.0

    try:
        popt, _ = curve_fit(sersic_profile, r_fit, s_fit,
                            p0=[guess_Sigma_e, guess_Re, guess_n], maxfev=2000)
        return (popt[0], popt[1], popt[2])
    except RuntimeError:
        logging.warning("fit_sersic: curve_fit did not converge.")
        return (np.nan, np.nan, np.nan)


def fit_exponential(rvals, sigma_vals):
    """Fit Exp => (Sigma0, h)."""
    mask = (rvals>0) & (sigma_vals>0)
    r_fit = rvals[mask]
    s_fit = sigma_vals[mask]
    if len(r_fit) < 5:
        logging.warning("fit_exponential: insufficient data.")
        return (np.nan, np.nan)

    guess_Sigma0 = np.max(s_fit)
    guess_h      = 2.0
    try:
        popt, _ = curve_fit(exponential_profile, r_fit, s_fit,
                            p0=[guess_Sigma0, guess_h], maxfev=2000)
        return (popt[0], popt[1])
    except RuntimeError:
        logging.warning("fit_exponential: curve_fit did not converge.")
        return (np.nan, np.nan)

print("Cell 4.2.1 done: Sersic & exponential fitting functions defined.")
```

```python
# =========================================
# Cell 4.2.2 — Run Fits for All Snapshots & Store Parameters
# =========================================

def run_fits_for_all_snapshots(all_profiles, save_results=SAVE_INTERMEDIATE_RESULTS):
    """
    For each snapshot in all_profiles, do both Sersic & exponential fits,
    store results in dicts => (time, paramA, paramB, ...).
    """
    sersic_fits = {}
    exp_fits    = {}

    for snap_idx in sorted(all_profiles.keys()):
        t_val, r_mid, Sigma = all_profiles[snap_idx]

        Se, Re, n_ = fit_sersic(r_mid, Sigma)
        S0, h_     = fit_exponential(r_mid, Sigma)

        sersic_fits[snap_idx] = (t_val, Se, Re, n_)
        exp_fits[snap_idx]    = (t_val, S0, h_)

        if np.isnan(Re) or np.isnan(n_):
            logging.warning(f"Sersic => snap {snap_idx} returned NaNs.")
        if np.isnan(S0) or np.isnan(h_):
            logging.warning(f"Exponential => snap {snap_idx} returned NaNs.")

        logging.info(f"Snap {snap_idx} at t={t_val:.1f} => Sersic(Re={Re:.2f}, n={n_:.2f}), Exp(h={h_:.2f})")

    if save_results and len(all_profiles)>0:
        os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
        # Save as .npz
        sersic_out = os.path.join(RESULTS_DIRECTORY, "sersic_fits.npz")
        exp_out    = os.path.join(RESULTS_DIRECTORY, "exp_fits.npz")

        sersic_dict = {}
        exp_dict    = {}

        for snap_idx in sersic_fits:
            t_val, Se, Re, n_ = sersic_fits[snap_idx]
            sersic_dict[f"Snap{snap_idx}_time"]   = np.array([t_val])
            sersic_dict[f"Snap{snap_idx}_SigmaE"] = np.array([Se])
            sersic_dict[f"Snap{snap_idx}_Re"]     = np.array([Re])
            sersic_dict[f"Snap{snap_idx}_n"]      = np.array([n_])

            t2, S0, h_ = exp_fits[snap_idx]
            exp_dict[f"Snap{snap_idx}_time"]   = np.array([t2])
            exp_dict[f"Snap{snap_idx}_Sigma0"] = np.array([S0])
            exp_dict[f"Snap{snap_idx}_h"]      = np.array([h_])

        np.savez(sersic_out, **sersic_dict)
        np.savez(exp_out, **exp_dict)
        logging.info(f"Sersic fits => {sersic_out}")
        logging.info(f"Exp fits => {exp_out}")

    return sersic_fits, exp_fits

# Example usage
sersic_fits, exp_fits = run_fits_for_all_snapshots(all_profiles)
print("Cell 4.2.2 done: best-fit Sersic & Exponential parameters computed.")
```

```python
# =========================================
# Cell 4.2.3 — Timeseries of Fit Parameters
# =========================================

def plot_fit_parameters_timeseries(sersic_fits, exp_fits, outfig="sersic_exp_timeseries.png"):
    """
    Plots Re, n (Sersic) & h (Exponential) vs. time. 3 subplots.
    """
    if not sersic_fits and not exp_fits:
        print("No fit data found.")
        return

    # gather
    s_keys = sorted(sersic_fits.keys())
    e_keys = sorted(exp_fits.keys())

    t_sersic = []
    Re_list  = []
    n_list   = []
    for k in s_keys:
        t_, Se, Re, n_ = sersic_fits[k]
        t_sersic.append(t_)
        Re_list.append(Re)
        n_list.append(n_)

    t_exp = []
    h_list = []
    for k in e_keys:
        t2, S0, h_ = exp_fits[k]
        t_exp.append(t2)
        h_list.append(h_)

    # arrays & sort
    t_sersic = np.array(t_sersic)
    Re_arr   = np.array(Re_list)
    n_arr    = np.array(n_list)
    idx_s    = np.argsort(t_sersic)
    t_sersic = t_sersic[idx_s]
    Re_arr   = Re_arr[idx_s]
    n_arr    = n_arr[idx_s]

    t_exp  = np.array(t_exp)
    h_arr  = np.array(h_list)
    idx_e  = np.argsort(t_exp)
    t_exp  = t_exp[idx_e]
    h_arr  = h_arr[idx_e]

    fig, axs = plt.subplots(3,1, figsize=(10,12))
    # Re
    axs[0].plot(t_sersic, Re_arr, '-', label='Re')
    axs[0].set_title("Sersic Re vs. Time")
    axs[0].set_xlabel("Time (Gyr)")
    axs[0].set_ylabel("Re (kpc)")

    # n
    axs[1].plot(t_sersic, n_arr, '-', color='orange', label='n')
    axs[1].set_title("Sersic n vs. Time")
    axs[1].set_xlabel("Time (Gyr)")
    axs[1].set_ylabel("n")

    # h
    axs[2].plot(t_exp, h_arr, '-', color='green', label='h')
    axs[2].set_title("Exponential h vs. Time")
    axs[2].set_xlabel("Time (Gyr)")
    axs[2].set_ylabel("h (kpc)")

    for ax in axs:
        ax.grid(True)
        ax.legend()

    plt.tight_layout()
    if outfig:
        os.makedirs(PLOTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(PLOTS_DIRECTORY, outfig)
        plt.savefig(outfile, dpi=150)
        print(f"Fit parameters timeseries saved -> {outfile}")

    plt.show()

# Example usage
plot_fit_parameters_timeseries(sersic_fits, exp_fits)
print("Cell 4.2.3 done: Re, n, h timeseries plotted.")
```

```python
# =========================================
# Cell 4.3.1 — Slope-based Tidal Truncation Analysis
# =========================================

def find_truncation_radius(rvals, sigma_vals, slope_threshold=-4.0):
    """
    Returns the first radius where the local slope of log(Sigma) vs. log(R)
    drops below slope_threshold. If none found, returns np.nan.
    """
    mask = (rvals>0) & (sigma_vals>0)
    r_ok = rvals[mask]
    s_ok = sigma_vals[mask]
    if len(r_ok) < 3:
        return np.nan

    logR = np.log10(r_ok)
    logS = np.log10(s_ok)
    slopes = []
    r_mid  = []

    for i in range(len(r_ok)-1):
        dlogS = logS[i+1] - logS[i]
        dlogR = logR[i+1] - logR[i]
        if abs(dlogR) < 1e-10:
            continue
        slope_i = dlogS / dlogR
        slopes.append(slope_i)
        r_mid.append(0.5*(r_ok[i] + r_ok[i+1]))

    slopes = np.array(slopes)
    r_mid  = np.array(r_mid)
    if len(slopes)==0:
        return np.nan

    idx = np.where(slopes < slope_threshold)[0]
    return np.nan if len(idx)==0 else r_mid[idx[0]]


def compute_tidal_truncation(all_profiles, slope_threshold=-4.0, save_results=SAVE_INTERMEDIATE_RESULTS):
    """
    For each snapshot's (r_mid, Sigma), find the radius of tidal truncation
    where slope < slope_threshold. Return times, R_trunc arrays.
    """
    snap_list = sorted(all_profiles.keys())
    times_t   = []
    rtrunc    = []

    for snap_idx in snap_list:
        t_val, r_mid, Sigma = all_profiles[snap_idx]
        R_cut = find_truncation_radius(r_mid, Sigma, slope_threshold)
        logging.info(f"Snap {snap_idx}: t={t_val:.1f} => R_trunc={R_cut:.2f}")
        times_t.append(t_val)
        rtrunc.append(R_cut)

    times_t = np.array(times_t)
    rtrunc  = np.array(rtrunc)
    idx     = np.argsort(times_t)
    times_t = times_t[idx]
    rtrunc  = rtrunc[idx]

    if save_results and len(times_t)>0:
        os.makedirs(RESULTS_DIRECTORY, exist_ok=True)
        outfn = os.path.join(RESULTS_DIRECTORY, "tidal_truncation_vs_time.npy")
        np.save(outfn, np.column_stack((times_t, rtrunc)))
        logging.info(f"Tidal truncation results => {outfn}")

    return times_t, rtrunc

# Run
times_trunc, Rtrunc_arr = compute_tidal_truncation(all_profiles, slope_threshold=-4.0)
print("Cell 4.3.1 done: slope-based tidal truncation analysis completed.")
```

```python
# =========================================
# Cell 4.3.2 — Plot Tidal Truncation Radius vs. Time
# =========================================

def plot_tidal_truncation_vs_time(times_trunc, Rtrunc_arr,
                                  times_jacobi=None, rjacobi_arr=None,
                                  slope_threshold=-4.0,
                                  outfig="tidal_truncation_vs_time.png"):
    """
    Plots R_trunc(t) from slope-based method. Optionally overlay Jacobi radius.
    """
    if len(times_trunc)==0:
        print("No tidal truncation data to plot.")
        return

    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(times_trunc, Rtrunc_arr, '-r', label='Tidal Truncation Radius')

    # overlay jacobi
    if (times_jacobi is not None) and (rjacobi_arr is not None) \
       and (len(times_jacobi)==len(rjacobi_arr)) and (len(times_jacobi)>0):
        idx_j = np.argsort(times_jacobi)
        t_j   = times_jacobi[idx_j]
        r_j   = rjacobi_arr[idx_j]
        ax.plot(t_j, r_j, '-b', alpha=0.6, label='Jacobi Radius')

    ax.set_xlabel("Time (Gyr)")
    ax.set_ylabel("Radius (kpc)")
    ax.set_title(f"Tidal Truncation vs. Time (slope<{slope_threshold})")
    ax.grid(True)
    ax.legend()

    if outfig:
        os.makedirs(PLOTS_DIRECTORY, exist_ok=True)
        outfile = os.path.join(PLOTS_DIRECTORY, outfig)
        plt.tight_layout()
        plt.savefig(outfile, dpi=150)
        print(f"Tidal truncation plot saved -> {outfile}")

    plt.show()

# Plot
plot_tidal_truncation_vs_time(times_trunc, Rtrunc_arr, times_jacobi, rjacobi_arr)
print("Cell 4.3.2 done: tidal truncation radius vs. time plot generated.")
```
